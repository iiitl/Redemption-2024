```python
label_encoder = LabelEncoder()                                                        #it is very important step in ml in which we convert text data into numerical data so that model can use that data for predictions
df['Credit_Score_encoded'] = label_encoder.fit_transform(df['Credit_Score'])
df['Credit_Mix_encoded'] = label_encoder.fit_transform(df['Credit_Mix'])
df['Payment_of_Min_Amount_encoded'] = label_encoder.fit_transform(df['Payment_of_Min_Amount'])
df['Payment_Behaviour_encoded'] = label_encoder.fit_transform(df['Payment_Behaviour'])
df.drop(columns=['Credit_Score', 'Credit_Mix', 'Payment_of_Min_Amount','Payment_Behaviour'], inplace=True)
```

```python
df.head()
```

```python
df.info() # this function is used before StandardScaler because Standard scaler require numerical value and info() helps to tell the datatype of column
```

```python
Scaler = StandardScaler()
X = df.drop(columns=['Credit_Score_encoded'], axis=1)#features        # standard Scaler function is used to make the range of the values in the column of same range
Y = df['Credit_Score_encoded']#labels
```

```python
standarized_Y = Scaler.fit_transform(Y.values.reshape(-1, 1))
print(standarized_Y)
```

```python
standardize_X = Scaler.fit_transform(X)
print(standardize_X)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3) # train_test_split is used to break the data in training and test data, in majority case split between training and test data is 80:20 so that it mostly trained on test data
```

```python
model = RandomForestClassifier() # after performing various models and calculating there accuracy score the Random forest Classfier has the best accuracy score
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
X_train_accuracy = accuracy_score(X_train_prediction,Y_train)
print(X_train_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
X_test_accuracy = accuracy_score(X_test_prediction, Y_test)
print(X_test_accuracy)
```

```python
conf_matrix = confusion_matrix(Y_test, X_test_prediction)            #confussion matrix gives the detailed analysis of the classification model
plt.figure(figsize=(10, 10))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap='Blues', cbar=False,
            linewidths=1, linecolor='black', square=True, mask=None)
plt.title('Confusion Matrix_test')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')

plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
