```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('/content/insurance (1).csv')
df.head()
```

```python
for column in df:
    plt.figure(figsize=(8, 6))
    sns.histplot(df[column], kde=True, color='red')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

```python
label_encoder = LabelEncoder()
df['sex_encoded'] = label_encoder.fit_transform(df['sex'])
df['region_encoded'] = label_encoder.fit_transform(df['region'])
df['smoker_encoded'] = label_encoder.fit_transform(df['smoker'])
df.drop(columns=['sex', 'region','smoker'], inplace=True)
```

```python
df.head()
```

```python
for column in df:
    plt.figure(figsize=(8, 6))
    sns.kdeplot(df[column], color='green', shade=True)
    plt.title(f'Density Plot of {column}')
    plt.xlabel(column)
    plt.ylabel('Density')
    plt.grid(True)
    plt.show()
```
