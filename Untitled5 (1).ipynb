```python
max_season_count = df['season'].value_counts().idxmax()
print(max_season_count)
```

```python
df['venue'].value_counts()
```

```python
venue_counts = df['venue'].value_counts()
venue_counts_nonzero = venue_counts[venue_counts > 0]
labels = venue_counts_nonzero.index
values = venue_counts_nonzero.values
plt.figure(figsize=(10, 6), dpi=100)
plt.bar(labels, values, color='skyblue')


plt.xticks(rotation=90)


plt.xlabel('Venue')
plt.ylabel('Count')
plt.title('Counts of Venues')


plt.tight_layout()
plt.show()
```
