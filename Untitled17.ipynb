```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from keras.layers import Dense, BatchNormalization, Dropout, LSTM
from keras.models import Sequential
from sklearn.metrics import mean_squared_error
import tensorflow as tf
from keras.models import load_model
import keras
import os
```

```python
df = pd.read_csv("/content/train.csv")
df.head()
```

```python
corr = df.corr()
```

```python
plt.figure(figsize=(30, 30))
sns.heatmap(corr, cbar=True, square=True, fmt='.2f', annot=True, annot_kws={'size': 10}, cmap='Blues', linewidths=0.5, linecolor='gray')
plt.title("Correlation Matrix", fontsize=16, fontweight='bold', pad=20)
plt.xticks(fontsize=10, fontweight='bold')
plt.yticks(fontsize=10, fontweight='bold')
plt.xlabel('Features', fontsize=12, fontweight='bold', labelpad=15)
plt.ylabel('Features', fontsize=12, fontweight='bold', labelpad=15)
plt.show()
```

```python
scaler = StandardScaler()
data_array = df.values
scaled_data_array = scaler.fit_transform(data_array)
df = pd.DataFrame(scaled_data_array, columns=df.columns)
```

```python
y = df['shares']
X = df.drop(['shares'], axis = 1)
```

```python
X = X.head(15000)
y = y.head(15000)
```

```python
X.shape
```

```python
model = Sequential()

model.add(Dense(128, input_shape=(X.shape[1],), activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(1, activation='linear'))

model.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python

model.fit(X, y, epochs=200, batch_size=32, validation_split=0.2)
```

```python
model.save('model.h5')
```
