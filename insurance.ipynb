```python
fig, axes = plt.subplots(nrows=len(df.columns), ncols=1, figsize=(30, 60))

for i, col in enumerate(df.columns):
    axes[i].hist(df[col], color='skyblue', edgecolor='black')
    axes[i].set_title(col)
    axes[i].set_xlabel('Value')
    axes[i].set_ylabel('Frequency')


plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dfe=['sex','smoker','region']
for col in dfe:
    df[col] = le.fit_transform(df[col])
```
