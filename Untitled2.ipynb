```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import zscore
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
df = pd.read_csv('/content/original data.csv')
df.head()
```

```python
plt.figure(figsize=(8, 8))
df = df.drop(columns=['Unnamed: 0'], errors='ignore')
sns.boxplot(data=df, orient="h", palette="Set2")
plt.xlim(-5, 5)
plt.title('Box Plot of Features')
plt.xlabel('Values')
plt.ylabel('Features')
plt.grid(True)
plt.show()
```

```python
def detect_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return (series < lower_bound) | (series > upper_bound)

outliers = df.select_dtypes(include=['number']).apply(detect_outliers)
outliers_melted = outliers.reset_index().melt(id_vars='index', var_name='Feature', value_name='Outlier')

plt.figure(figsize=(12, 8))
sns.scatterplot(data=outliers_melted, x='index', y='Feature', hue='Outlier', hue_order=[True, False], palette={True: 'red', False: 'black'}, marker='x', s=100, alpha=0.5)
plt.title('Outliers Detection using IQR Method')
plt.xlabel('Data Points')
plt.ylabel('Features')
plt.grid(True)
plt.legend(title='Outlier', loc='upper right', labels=['Non Outliner', 'Outlier'])
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
threshold_z = 2
outlier_indices = []
for column in df.columns:
    if column != 'label':

        z = np.abs((df[column] - df[column].mean()) / df[column].std())


        column_outliers = np.where(z > threshold_z)[0]


        outlier_indices.extend(column_outliers)


outlier_indices = np.unique(outlier_indices)
no_outliers = df.drop(outlier_indices)
plt.figure(figsize=(8, 8))
for column in df.columns:
    if column != 'label':
        plt.scatter(df.index, df[column], color='black', label='Data Points')
        plt.scatter(df.index[outlier_indices], df[column].iloc[outlier_indices], color='red', marker='x', label='Outliers')
        plt.xlabel('Data Points')
        plt.ylabel(column)
        plt.title(f'Scatter Plot of {column} with Outliers')
        plt.legend()
        plt.grid(True)
        plt.show()
```
